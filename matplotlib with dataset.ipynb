```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df= pd.read_csv("D://dataset//data.csv")

df
```

```python
x = df["Duration"]
y= df["Calories"]


plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker="o")
plt.show()
```

```python

x = df["Duration"]
y= df["Calories"]


plt.plot(x,y)

plt.title("workout comparision",loc="left")
plt.xlabel("Duration")
plt.ylabel("Calories")

plt.show()
```

## Set Font Properties for Title and Labels
You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels

```python
#scatter

plt.scatter(x,y)
plt.show()
```

```python
plt.bar(x,y,width=10,color = "hotpink")
```

```python
plt.hist(x)
plt.show()
```

## Matplotlib Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```
